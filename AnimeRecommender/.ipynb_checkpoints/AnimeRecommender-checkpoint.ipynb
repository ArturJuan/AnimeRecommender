```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
```

```python
anime = pd.read_csv('archive/anime.csv')
rating = pd.read_csv('archive/rating.csv')
```

```python
anime.head(10)
```

```python
rating.head()
```

```python
rating_groupby = rating.groupby('anime_id', as_index=False, sort=False).agg({'user_id':'first', 'rating':'mean'})
```

```python
df = anime.merge(rating_groupby, on='anime_id')
df.drop('rating_x', inplace=True, axis=1)
```
